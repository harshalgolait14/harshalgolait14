```python
image = cv2.imread(r"F:\IMARTICUS\Data\OpenCV\WhatsApp Image 2022-02-07 at 9.55.36 AM.jpeg")
```

```python
face_cascade = cv2.CascadeClassifier(r"F:\IMARTICUS\Data\Haarcascade\Haarcascade\haarcascade_frontalface_alt.xml")
```

```python
def  face_detection(img):
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img) # this face_rects is just x and y postion and height & breadth of rectangle
    
    for ( x, y , w, h) in face_rects:
        cv2.rectangle(face_img  ,(x, y) , (x+w , y+h) , (255, 0, 0) , 3)  # drawing the red rectangle on face
    return face_img
```

```python
image_with_rect = face_detection(image)
```

```python
plt.figure(figsize= (10 , 10))
plt.imshow(image_with_rect)
```

```python
def  eye_detection(img):
    face_img = img.copy()
    
    eye_rects = eye_cascade.detectMultiScale(face_img) # this face_rects is just x and y postion and height & breadth of rectangle
    
    for ( x, y , w, h) in eye_rects:
        cv2.rectangle(face_img  ,(x, y) , (x+w , y+h) , (255, 0, 0) , 2)  # drawing the red rectangle on face
    return face_img
```
